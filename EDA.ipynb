```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Concrete_Data.csv")
df.head(5)
```

```python
df.info()
```

```python
result=round(df.describe(),2)
result
```

```python
sns.pairplot(df)
```

```python
plt.figure(figsize=[15,7])
sns.heatmap(df.corr(), annot=True, cmap='Blues')
```

```python
plt.figure(figsize=[15,6])
sns.scatterplot(y="CCStr", x="Cement", hue="Water",size="Superplastic", data=df, sizes=(50, 300))
```

```python
col=["Cement","Slag","Fly_Ash","Water","Superplastic","Coarse_Aggr","Fine_Aggr","Age","CCStr"]
for i in range(len(col)):
    sns.set(font_scale = 1.5)
    sns.displot(df, x=col[i], binwidth=5)
```
